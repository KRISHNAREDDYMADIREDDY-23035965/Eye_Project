```python
# Import necessary libraries
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Path to the folder containing images (organized in subfolders for each class)
folder_path = "/content/drive/MyDrive/dataset"

# Load the full dataset
sample_ds = image_dataset_from_directory(
    folder_path,
    image_size=(128, 128),
    batch_size=1,
    shuffle=False  # We don’t need to shuffle since we want all images
)

# Function to compute brightness of a single image
def compute_brightness(img_tensor):
    img_array = img_tensor.numpy().astype("uint8")
    img_gray = cv2.cvtColor(img_array[0], cv2.COLOR_RGB2GRAY)
    return np.mean(img_gray)

# Collect brightness values for all 2000 images
brightness_values = []
for images, _ in sample_ds:
    brightness = compute_brightness(images)
    brightness_values.append(brightness)

# Compute statistics
mean_brightness = np.mean(brightness_values)
median_brightness = np.median(brightness_values)
std_brightness = np.std(brightness_values)

# Plot histogram of brightness values with statistical lines
plt.figure(figsize=(8, 5))
plt.hist(brightness_values, bins=30, color="orange", edgecolor="black")

plt.axvline(mean_brightness, color='blue', linestyle='dashed', linewidth=1.5, label=f"Mean: {mean_brightness:.2f}")
plt.axvline(median_brightness, color='green', linestyle='dashed', linewidth=1.5, label=f"Median: {median_brightness:.2f}")
plt.axvline(mean_brightness + std_brightness, color='red', linestyle='dotted', linewidth=1, label=f"+1 SD: {(mean_brightness + std_brightness):.2f}")
plt.axvline(mean_brightness - std_brightness, color='red', linestyle='dotted', linewidth=1, label=f"-1 SD: {(mean_brightness - std_brightness):.2f}")

plt.title("Brightness Distribution of Images")
plt.xlabel("Average Brightness (0-255)")
plt.ylabel("Number of Images")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#-----------------custom Cnn Model--------------------
# --- Importing Required Libraries ---
import tensorflow as tf
from tensorflow import keras  # High-level Keras API for building models
from tensorflow.keras import layers  # Layers module for model building
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc  # Evaluation metrics
from sklearn.utils.class_weight import compute_class_weight  # For handling class imbalance
import seaborn as sns

data_path = "/content/drive/MyDrive/dataset"  # Path to the dataset folder
batch_size = 64  # Batch size for training
img_size = (128, 128)  # Target image size (height, width)
seed = 60

# --- Load Dataset Helper Function ---
def load_dataset(subset, split=0.3):

    return tf.keras.utils.image_dataset_from_directory(
        data_path,
        validation_split=split,
        subset=subset,  # Either "training" or "validation"
        seed=seed,
        image_size=img_size,
        batch_size=batch_size,
        label_mode="categorical"#One-hot
    )

# --- Load Training and Validation Datasets ---
train_ds = load_dataset("training")
val_ds = load_dataset("validation")
class_names = train_ds.class_names
num_classes = len(class_names)

# --- Preprocessing Function ---
def preprocess(image, label, augment=False):
    # Normalize pixel values
    image = tf.cast(image, tf.float32) / 255.0
    # Resize to standard size
    image = tf.image.resize(image, img_size)
    # Apply data augmentation if enabled
    if augment:
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_flip_up_down(image)
        image = tf.image.random_brightness(image, 0.1)
        image = tf.image.random_contrast(image, 0.8, 1.2)
        image = tf.image.random_saturation(image, 0.8, 1.2)
        image = tf.image.random_hue(image, 0.05)
    return image, label

# Apply preprocessing to datasets
train_ds = train_ds.map(lambda x, y: preprocess(x, y, augment=True)).prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.map(lambda x, y: preprocess(x, y)).prefetch(tf.data.AUTOTUNE)

# --- Compute Class Weights (for Imbalanced Data) ---
y_labels = []
# Unbatch the dataset to get individual labels
for _, labels in train_ds.unbatch():
    y_labels.append(np.argmax(labels.numpy()))  # Convert one-hot to integer labels

# Compute balanced class weights
class_weights = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_labels)
class_weights = dict(enumerate(class_weights))  # Convert to dictionary format

# --- Build CNN Model ---
def build_model(input_shape=(128, 128, 3), num_classes=4):
    l2 = keras.regularizers.l2(1e-4)  # L2 regularization to prevent overfitting
    inputs = keras.Input(shape=input_shape)  # Input layer

    # Block 1
    x = layers.Conv2D(32, 3, padding="same", kernel_regularizer=l2)(inputs)
    x = layers.LeakyReLU()(x)
    x = layers.Conv2D(64, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)

    # Block 2
    x = layers.Conv2D(64, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.3)(x)  # Dropout to reduce overfitting

    # Block 3
    x = layers.Conv2D(128, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.4)(x)

    # Block 4
    x = layers.Conv2D(256, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.5)(x)

    # Fully Connected Layers
    x = layers.Flatten()(x) # Converts the output of the last convolutional layer into a flat 1D vector.
    x = layers.Dense(512, kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.5)(x)

    # Output Layer
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs)  # Define model structure

    # Compile the model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.0005),  # Adam optimizer
        loss=keras.losses.CategoricalCrossentropy(label_smoothing=0.1),  # Softmax loss with label smoothing
        metrics=["accuracy"]  # Track accuracy
    )
    return model

# --- Train the Model ---
model = build_model(input_shape=(128, 128, 3), num_classes=num_classes)

# Define callbacks
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),  # Stop early if no improvement
    keras.callbacks.ModelCheckpoint("best_modelh.keras", save_best_only=True, monitor="val_accuracy"),  # Save best model
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.3, patience=4, min_lr=1e-6)  # Reduce LR on plateau
]

model.summary()  # Print model architecture
# Train the model
history = model.fit(train_ds, validation_data=val_ds, epochs=80, callbacks=callbacks, class_weight=class_weights)
model.save("Custom_Cnn_model.keras")

# --- Evaluation ---
val_loss, val_acc = model.evaluate(val_ds, verbose=0)
print(f"\nTraining Accuracy: {history.history['accuracy'][-1]*100:.2f}%")  # Final training accuracy
print(f"Validation Accuracy: {val_acc*100:.2f}%")  # Final validation accuracy

# --- Accuracy Plot ---
plt.figure(figsize=(8, 6))
plt.plot(history.history["accuracy"], label="Training")  # Training accuracy
plt.plot(history.history["val_accuracy"], label="Validation", linestyle="--")  # Validation accuracy
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Loss Plot ---
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training")  # Training loss
plt.plot(history.history["val_loss"], label="Validation", linestyle="--")  # Validation loss
plt.title("Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Predictions on Validation Data ---
y_true, y_pred, y_prob = [], [], []
for x, y in val_ds:
    pred = model.predict(x)  # Model predictions
    y_true.extend(np.argmax(y.numpy(), axis=1))  # Ground truth labels
    y_pred.extend(np.argmax(pred, axis=1))  # Predicted class labels
    y_prob.extend(pred)  # Prediction probabilities

# Convert to numpy arrays
y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_prob = np.array(y_prob)

# --- Classification Report ---
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))  # Precision, recall, f1-score

# --- Confusion Matrix ---
cm = confusion_matrix(y_true, y_pred)  # Create confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)  # Heatmap plot
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# --- ROC Curve for Multi-class ---
plt.figure(figsize=(8, 6))
for i in range(num_classes):
    fpr, tpr, _ = roc_curve(y_true == i, y_prob[:, i])  # Calculate FPR and TPR for class i
    auc_score = auc(fpr, tpr)  # Compute AUC
    plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC={auc_score:.2f})")  # Plot ROC curve for class i

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.title("ROC Curve")
plt.xlabel("FPR")#TPR (True Positive Rate)
plt.ylabel("TPR")#FPR (False Positive Rate)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image

# Set image path
img_path = "/content/drive/MyDrive/dataset/diabetic_retinopathy/1000_left.jpeg"

# Function to load and preprocess image
def preprocess_img(img_path):
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array /= 255.0  # Normalize
    return img_array, img

# Load the trained model
model = keras.models.load_model("best_modelh.keras")

# Find the last Conv2D layer dynamically
def get_last_conv_layer(model):
    for layer in reversed(model.layers):
        if isinstance(layer, layers.Conv2D):
            return layer.name
    raise ValueError("No convolutional layer found.")

last_conv_layer_name = get_last_conv_layer(model)
print(f"Using last conv layer: {last_conv_layer_name}")

# Generate Grad-CAM heatmap
def make_gradcam_heatmap(img_array, model, last_conv_layer_name):
    grad_model = Model(
        inputs=model.input,
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_output, predictions = grad_model(img_array)
        class_index = tf.argmax(predictions[0])  # Get predicted class index
        loss = predictions[:, class_index]

    grads = tape.gradient(loss, conv_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_output = conv_output[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_output), axis=-1)

    # Fix: Apply ReLU and normalize properly
    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap) if np.max(heatmap) != 0 else 1
    return heatmap

# Generate Grad-CAM heatmap
img_array, orig_img = preprocess_img(img_path)
heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name)

# Overlay Grad-CAM heatmap on original image with improved blending
def overlay_gradcam(img_path, heatmap, alpha=0.4):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (128, 128))

    # Fix: Use INTER_CUBIC for smooth interpolation
    heatmap = cv2.resize(heatmap, (128, 128), interpolation=cv2.INTER_CUBIC)
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # Fix: Adjust alpha blending
    overlay = cv2.addWeighted(img, 1 - alpha, heatmap, alpha, 0)
    return overlay

overlayed_img = overlay_gradcam(img_path, heatmap)

# Plot results
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(orig_img)
ax[0].set_title("Original Image")
ax[0].axis("off")
ax[1].imshow(heatmap, cmap="jet")
ax[1].set_title("Grad-CAM Heatmap")
ax[1].axis("off")
ax[2].imshow(overlayed_img)
ax[2].set_title("Overlay")
ax[2].axis("off")
plt.show()
```

```python
# --- Imports ---
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
from sklearn.utils.class_weight import compute_class_weight

# --- Dataset Settings ---
data_path = "/content/drive/MyDrive/dataset"
img_size = (128, 128)
batch_size = 64
seed = 60
epochs = 30

# --- Load Dataset ---
def load_dataset(subset):
    return tf.keras.utils.image_dataset_from_directory(
        data_path,
        validation_split=0.3,
        subset=subset,
        seed=seed,
        image_size=img_size,
        batch_size=batch_size,
        label_mode="categorical"
    )

train_ds = load_dataset("training")
val_ds = load_dataset("validation")
class_names = train_ds.class_names
num_classes = len(class_names)

# --- Preprocessing Function ---
def preprocess(image, label, augment=False):
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.image.resize(image, img_size)
    if augment:
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_brightness(image, 0.1)
    return image, label

train_ds = train_ds.map(lambda x, y: preprocess(x, y, augment=True)).prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.map(preprocess).prefetch(tf.data.AUTOTUNE)

# --- Compute Class Weights ---
y_labels = [np.argmax(y.numpy()) for _, y in train_ds.unbatch()]
class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(y_labels), y=y_labels)
class_weights_dict = {i: w for i, w in enumerate(class_weights)}

# --- Custom CNN Model ---
def build_custom_cnn(input_shape=(128, 128, 3), num_classes=4):
    l2 = keras.regularizers.l2(1e-4)
    inputs = keras.Input(shape=input_shape)

    x = layers.Conv2D(32, 3, padding="same", kernel_regularizer=l2)(inputs)
    x = layers.LeakyReLU()(x)
    x = layers.Conv2D(64, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(64, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(128, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Conv2D(256, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.5)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(512, kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.5)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs)
    model.compile(
        optimizer=keras.optimizers.Adam(0.0005),
        loss=keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
        metrics=["accuracy"]
    )
    return model

# --- VGG16 Model ---
def build_vgg16(input_shape=(128, 128, 3), num_classes=4):
    base_model = keras.applications.VGG16(include_top=False, weights="imagenet", input_shape=input_shape)
    base_model.trainable = False

    inputs = keras.Input(shape=input_shape)
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs, outputs)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-4),
        loss=keras.losses.CategoricalCrossentropy(),
        metrics=["accuracy"]
    )
    return model

# --- MobileNetV2 Model ---
def build_mobilenet(input_shape=(128, 128, 3), num_classes=4):
    base_model = keras.applications.MobileNetV2(include_top=False, weights="imagenet", input_shape=input_shape)
    base_model.trainable = False

    inputs = keras.Input(shape=input_shape)
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs, outputs)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-4),
        loss=keras.losses.CategoricalCrossentropy(),
        metrics=["accuracy"]
    )
    return model

# --- Train All Models for 30 Epochs ---
print("Training Custom CNN...")
custom_model = build_custom_cnn()
history_custom = custom_model.fit(train_ds, validation_data=val_ds, epochs=epochs, class_weight=class_weights_dict, verbose=2)

print("Training VGG16...")
vgg_model = build_vgg16()
history_vgg = vgg_model.fit(train_ds, validation_data=val_ds, epochs=epochs, class_weight=class_weights_dict, verbose=2)

print("Training MobileNetV2...")
mobile_model = build_mobilenet()
history_mobile = mobile_model.fit(train_ds, validation_data=val_ds, epochs=epochs, class_weight=class_weights_dict, verbose=2)

# --- Plot Accuracy Comparison ---
plt.figure(figsize=(10, 6))
plt.plot(history_custom.history['val_accuracy'], label='Custom CNN')
plt.plot(history_vgg.history['val_accuracy'], label='VGG16')
plt.plot(history_mobile.history['val_accuracy'], label='MobileNetV2')
plt.title("Validation Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#-----------------custom Cnn Model--------------------
# --- Importing Required Libraries ---
import tensorflow as tf
from tensorflow import keras  # High-level Keras API for building models
from tensorflow.keras import layers  # Layers module for model building
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc  # Evaluation metrics
from sklearn.utils.class_weight import compute_class_weight  # For handling class imbalance
import seaborn as sns

data_path = "/content/drive/MyDrive/dataset"  # Path to the dataset folder
batch_size = 64  # Batch size for training
img_size = (224, 224)  # Target image size (height, width)
seed = 60

# --- Load Dataset Helper Function ---
def load_dataset(subset, split=0.2):

    return tf.keras.utils.image_dataset_from_directory(
        data_path,
        validation_split=split,
        subset=subset,  # Either "training" or "validation"
        seed=seed,
        image_size=img_size,
        batch_size=batch_size,
        label_mode="categorical"#One-hot
    )

# --- Load Training and Validation Datasets ---
train_ds = load_dataset("training")
val_ds = load_dataset("validation")
class_names = train_ds.class_names
num_classes = len(class_names)

# --- Preprocessing Function ---
def preprocess(image, label, augment=False):
    # Normalize pixel values
    image = tf.cast(image, tf.float32) / 255.0
    # Resize to standard size
    image = tf.image.resize(image, img_size)
    # Apply data augmentation if enabled
    if augment:
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_flip_up_down(image)
        image = tf.image.random_brightness(image, 0.1)
        image = tf.image.random_contrast(image, 0.8, 1.2)
        image = tf.image.random_saturation(image, 0.8, 1.2)
        image = tf.image.random_hue(image, 0.05)
    return image, label

# Apply preprocessing to datasets
train_ds = train_ds.map(lambda x, y: preprocess(x, y, augment=True)).prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.map(lambda x, y: preprocess(x, y)).prefetch(tf.data.AUTOTUNE)

# --- Compute Class Weights (for Imbalanced Data) ---
y_labels = []
# Unbatch the dataset to get individual labels
for _, labels in train_ds.unbatch():
    y_labels.append(np.argmax(labels.numpy()))  # Convert one-hot to integer labels

# Compute balanced class weights
class_weights = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_labels)
class_weights = dict(enumerate(class_weights))  # Convert to dictionary format

# --- Build CNN Model ---
def build_model(input_shape=(224, 224, 3), num_classes=4):
    l2 = keras.regularizers.l2(1e-4)  # L2 regularization to prevent overfitting
    inputs = keras.Input(shape=input_shape)  # Input layer

    # Block 1
    x = layers.Conv2D(32, 3, padding="same", kernel_regularizer=l2)(inputs)
    x = layers.LeakyReLU()(x)
    x = layers.Conv2D(64, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)

    # Block 2
    x = layers.Conv2D(64, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.3)(x)  # Dropout to reduce overfitting

    # Block 3
    x = layers.Conv2D(128, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.4)(x)

    # Block 4
    x = layers.Conv2D(256, 3, padding="same", kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.5)(x)

    # Fully Connected Layers
    x = layers.Flatten()(x) # Converts the output of the last convolutional layer into a flat 1D vector.
    x = layers.Dense(512, kernel_regularizer=l2)(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.5)(x)

    # Output Layer
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = keras.Model(inputs, outputs)  # Define model structure

    # Compile the model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.0005),  # Adam optimizer
        loss=keras.losses.CategoricalCrossentropy(label_smoothing=0.1),  # Softmax loss with label smoothing
        metrics=["accuracy"]  # Track accuracy
    )
    return model

# --- Train the Model ---
model = build_model(input_shape=(224, 224, 3), num_classes=num_classes)

# Define callbacks
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True),  # Stop early if no improvement
    keras.callbacks.ModelCheckpoint("best_modelh.keras", save_best_only=True, monitor="val_accuracy"),  # Save best model
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.3, patience=4, min_lr=1e-6)  # Reduce LR on plateau
]

model.summary()  # Print model architecture
# Train the model
history = model.fit(train_ds, validation_data=val_ds, epochs=50, callbacks=callbacks, class_weight=class_weights)
model.save("Custom_Cnn_model.keras")

# --- Evaluation ---
val_loss, val_acc = model.evaluate(val_ds, verbose=0)
print(f"\nTraining Accuracy: {history.history['accuracy'][-1]*100:.2f}%")  # Final training accuracy
print(f"Validation Accuracy: {val_acc*100:.2f}%")  # Final validation accuracy

# --- Accuracy Plot ---
plt.figure(figsize=(8, 6))
plt.plot(history.history["accuracy"], label="Training")  # Training accuracy
plt.plot(history.history["val_accuracy"], label="Validation", linestyle="--")  # Validation accuracy
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Loss Plot ---
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training")  # Training loss
plt.plot(history.history["val_loss"], label="Validation", linestyle="--")  # Validation loss
plt.title("Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Predictions on Validation Data ---
y_true, y_pred, y_prob = [], [], []
for x, y in val_ds:
    pred = model.predict(x)  # Model predictions
    y_true.extend(np.argmax(y.numpy(), axis=1))  # Ground truth labels
    y_pred.extend(np.argmax(pred, axis=1))  # Predicted class labels
    y_prob.extend(pred)  # Prediction probabilities

# Convert to numpy arrays
y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_prob = np.array(y_prob)

# --- Classification Report ---
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))  # Precision, recall, f1-score

# --- Confusion Matrix ---
cm = confusion_matrix(y_true, y_pred)  # Create confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)  # Heatmap plot
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# --- ROC Curve for Multi-class ---
plt.figure(figsize=(8, 6))
for i in range(num_classes):
    fpr, tpr, _ = roc_curve(y_true == i, y_prob[:, i])  # Calculate FPR and TPR for class i
    auc_score = auc(fpr, tpr)  # Compute AUC
    plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC={auc_score:.2f})")  # Plot ROC curve for class i

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.title("ROC Curve")
plt.xlabel("FPR")#TPR (True Positive Rate)
plt.ylabel("TPR")#FPR (False Positive Rate)
plt.legend()
plt.tight_layout()
plt.show()
```
